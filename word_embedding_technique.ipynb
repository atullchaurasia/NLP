```python
import tensorflow as tf
import warnings
warnings.filterwarnings("ignore")
print(tf.__version__)
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences = ['lion is the king of the jungle',
             'I like chicken tikka',
             'atul is a tech enthusiast', 
             'i am a good developer',
             'birds fly in the sky',
             'glass of juice', 
             'glass of milk']
```

```python
voc_size = 500 #vocabulary size
```

```python
onehot_rep = [one_hot(words, voc_size) for words in sentences]
onehot_rep
```

```python
import numpy as np
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length = 8
embedded_docs = pad_sequences(onehot_rep, padding='pre', maxlen=sent_length)
embedded_docs
```

```python
embedding_dim= 10
```

```python
model = Sequential()
model.add(Embedding(input_dim = voc_size, output_dim=embedding_dim,input_length=sent_length ))
model.build(input_shape=(None, sent_length))
model.compile(optimizer='adam', loss= 'mse')
```

```python
model.summary()
```

```python
print(model.predict(embedded_docs))
```
